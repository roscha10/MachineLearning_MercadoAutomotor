### Variable: car_ID - Identificación Única de Vehículo

Dentro de la columna 'CarName', se encuentran tanto la marca como el modelo de cada vehículo. Esta información es valiosa para nuestro análisis. Sin embargo, para trabajar de forma más efectiva, realizaremos una operación para separar esta columna en dos variables distintas: 'Marca' y 'Modelo'. Este paso nos brindará mayor facilidad y claridad al explorar y analizar los datos relacionados con los vehículos.

Después de segmentar la información en 'Marca' y 'Modelo', nos enfocaremos en garantizar que no haya datos incorrectos o mal escritos. 

A continuación, avanzaremos con la separación y examen de los valores únicos presentes en las nuevas variables 'Marca' y 'Modelo'

```python
#modificamos las marcas mal escritas a traves de un diccionario
data[['Marca', 'Modelo']] = data['CarName'].str.split(' ', n=1, expand=True)
```

```python
data.head()
```

```python
# Obtener los valores únicos en la columna 'Marca'
valores_unicos_marca = data['Marca'].unique()

# Mostrar los valores únicos
print("Valores únicos en la columna 'Marca':")
print(valores_unicos_marca)
```

En la exploración de los valores únicos en la columna 'Marca', observamos que algunos valores están mal escritos o tienen variaciones en la escritura que queremos corregir para mantener la consistencia. Por ejemplo, encontramos 'maxda' en lugar de 'mazda', 'Nissan' en lugar de 'nissan', 'porcshce' en lugar de 'porsche', 'toyouta' en lugar de 'toyota', 'vokswagen' en lugar de 'vw', y 'volkswagen' en lugar de 'vw'. Para abordar esto, utilizaremos un diccionario que asocie estos valores incorrectos con las correcciones apropiadas. Posteriormente, aplicaremos este diccionario para corregir estos errores y asegurar que nuestra columna 'Marca' esté uniformemente escrita.

```python
# Definición del diccionario de correcciones
correcciones_marca = {
    'maxda': 'mazda',
    'Nissan': 'nissan',
    'porcshce': 'porsche',
    'toyouta': 'toyota',
    'vokswagen': 'vw',
    'volkswagen': 'vw',
    'alfa-romero': 'alfa-romeo'  # Corrección específica para 'alfa-romero'
}

# Aplicación de las correcciones
data['Marca'] = data['Marca'].replace(correcciones_marca)

```

```python
# Obtener los valores únicos en la columna 'Marca'
valores_unicos_corregidos_marca = data['Marca'].unique()

# Mostrar los valores únicos
print("Valores únicos en la columna 'Marca':")
print(valores_unicos_corregidos_marca)
print('\nvalores unicos', len(valores_unicos_corregidos_marca))
```

### Transformación a Números: 'cylindernumber' y 'doornumber'

En nuestro conjunto de datos, tenemos dos variables, 'cylindernumber' y 'doornumber', que representan la cantidad de cilindros y puertas de cada vehículo, respectivamente. Estas variables están actualmente codificadas en formato de texto. Para facilitar su utilización en análisis y modelos, vamos a transformarlas en datos numéricos asignando valores numéricos a cada categoría. Por ejemplo, convertiremos 'four' a 4 y 'six' a 6 para 'cylindernumber'. Realizaremos un proceso similar para 'doornumber'. Esto nos permitirá trabajar de forma más efectiva con estos datos en futuras operaciones analíticas.

```python
# Mapeo de valores para 'cylindernumber'
tranformar_cylindernumber = {'four': 4, 'six': 6, 'five': 5, 'eight': 8, 'two': 2, 'twelve': 12, 'three': 3}
data['cylindernumber'] = data['cylindernumber'].map(tranformar_cylindernumber)

# Mapeo de valores para 'doornumber'
transformar_doornumber = {'two': 2, 'four': 4}
data['doornumber'] = data['doornumber'].map(transformar_doornumber)
```

```python
data[['cylindernumber', 'doornumber']].head(4)
```

Una vez analizadas las variables categóricas y numéricas, procedemos a abordar la gestión de los valores ausentes. Utilizaremos el método isna() para identificar la presencia de valores faltantes en cada columna y luego aplicaremos el método sum() para calcular el numero de valores ausentes en cada una de ellas.

```python
data.isna().sum()
```

Hemos identificado que en nuestra base de datos existen dos registros con información faltante en la columna 'Modelo'. Estos vehículos están asociados con la marca 'Subaru'. Para abordar esta situación, decidimos asignarles el modelo 'Desconocido', proporcionando así un valor representativo y uniforme para estos casos.

```python
# Filtrar las filas donde la columna 'Modelo' está vacía
modelos_vacios = data[data['Modelo'].isnull() | (data['Modelo'] == '')]

# Mostrar estos registros solo si hay filas con modelo vacío
if not modelos_vacios.empty:
    print("Registros con valor vacío en la columna 'Modelo':")
    print(modelos_vacios)
else:
    print("No hay registros con valor vacío en la columna 'Modelo'.")

```

```python
# Reemplazar valores faltantes en 'Modelo' con 'Desconocido'
data['Modelo'].fillna('Desconocido', inplace=True)
```

```python
data['Modelo'].isna().sum()
```

# Correlaciones entre Variables Categóricas

En esta sección, exploraremos las relaciones y asociaciones entre las variables categóricas de nuestro conjunto de datos. El análisis de correlación entre estas variables nos brinda información valiosa para comprender cómo distintas características interdependientes pueden influir en el precio de los automóviles. Utilizaremos técnicas especializadas como la `tabla de contingencia` y `el coeficiente de correlación de Cramer` para evaluar la fuerza y dirección de estas relaciones.

Comprender las interconexiones entre las características de un vehículo es fundamental para optimizar estrategias de marketing, personalizar ofertas, refinar estrategias de precios y optimizar la gama de productos en la industria automotriz. Además, este análisis nos permitirá identificar relaciones relevantes y potencialmente estratégicas entre variables categóricas, proporcionando insights valiosos para la toma de decisiones empresariales.

```python
from scipy.stats import chi2_contingency

### Función para calcular el coeficiente de Cramer's 
def calculate_cramers_v(confusion_matrix):
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k - 1) * (r - 1)) / (n - 1))
    rcorr = r - ((r - 1)**2) / (n - 1)
    kcorr = k - ((k - 1)**2) / (n - 1)
    return np.sqrt(phi2corr / min((kcorr - 1), (rcorr - 1)))


### DataFrame para almacenar los coeficientes de Cramer's 
cramer_matrix = pd.DataFrame(np.zeros((len(categorical_vars), 
                                       len(categorical_vars))),
                             columns=categorical_vars,
                             index=categorical_vars)

### Calcular coeficientes de Cramer's 
for var1 in categorical_vars:
    for var2 in categorical_vars:
        if var1 != var2:
            confusion_matrix = pd.crosstab(data[var1], data[var2])
            cramer_value = calculate_cramers_v(confusion_matrix.values)
            cramer_matrix.loc[var1, var2] = cramer_value

### Visualizar la matriz de coeficientes de Cramer's 
plt.figure(figsize=(8, 6))
sns.heatmap(cramer_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Coeficientes de Cramer\'s V entre Variables Categóricas')
plt.show()
```

# Preprocesamiento de Variables Categóricas para Modelos Predictivos

En la preparación de nuestras variables categóricas para los modelos predictivos, aplicaremos dos técnicas fundamentales: **Label Encoding** y **One-Hot Encoding**.

### Label Encoding:

Utilizaremos el Label Encoding para transformar la variable `symboling`, que representa el nivel de riesgo de seguro. Esta técnica asignará un número único a cada categoría en función de su posición en una escala ordinal. En este caso, asignaremos números a los diferentes niveles de riesgo, permitiendo que el modelo interprete su relación de manera adecuada.

### One-Hot Encoding:

Por otro lado, aplicaremos One-Hot Encoding a las variables de la lista `categorical_vars` (excluyendo `symboling`). Con One-Hot Encoding, crearemos nuevas columnas binarias para cada categoría en estas variables, indicando su presencia o ausencia. Este enfoque es crucial para evitar que el modelo asuma relaciones de orden entre las categorías y garantizar una representación precisa de estas variables en nuestro análisis.

Estas técnicas nos permitirán representar de manera efectiva nuestras variables categóricas, asegurando que los modelos puedan interpretarlas adecuadamente y capturar relaciones significativas. Continuemos aplicando estos conceptos en nuestro conjunto de datos para prepararlo óptimamente para nuestros futuros modelos predictivos.


```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

### (Label Encoding)
label_encoder = LabelEncoder()
data['symboling_encoded'] = label_encoder.fit_transform(data['symboling'])

### (One-Hot Encoding)
data = pd.get_dummies(data, columns=['fueltype', 'aspiration', 'carbody', 'drivewheel', 
                                    'enginelocation', 'enginetype', 'fuelsystem', 'Marca','doornumber','cylindernumber'], 
                      drop_first=True)
```

# Validación Cruzada para Modelos de Regresión

En este análisis, se explora la aplicación de técnicas de validación cruzada en un modelo de regresión. La validación cruzada es esencial en el proceso de evaluación de modelos de aprendizaje automático, ya que proporciona una evaluación más confiable del rendimiento del modelo.

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

# Creamos la funcion de Validación cruzada
def regresion(x_train, x_test, y_train, y_test):

    # Crear y entrenar el modelo
    linear_model = LinearRegression()
    linear_model.fit(x_train, y_train)

    score = linear_model.score(x_train, y_train)
 
    print("Metrica del modelo sin cross validation", score)
    
    kf = KFold(n_splits=7, shuffle=True, random_state=68)
 
    cv_scores = cross_val_score(linear_model, x_train, y_train, cv=kf)
    
    cv_scores = [round(num, 3) for num in cv_scores]
    print('Se obtienen los siguientes coeficientes de determinación con cross validation:')
    print(cv_scores)
    print(f'Max R-Squared: {max(cv_scores)}')
    print(f'Min R-Squared: {min(cv_scores)}')
    print('Promedio R-Squared: {:.3f}'.format(np.mean(cv_scores)))
    print('Desviación Estándar: {:.3f}'.format(np.std(cv_scores)))
    print(f'Intervalo de confianza 95%: {np.quantile(cv_scores, [0.025, 0.975])}', '\n')
 
    y_pred = linear_model.predict(x_test)
    print('Metricas en Test')
    print('Test MAE:', metrics.mean_absolute_error(y_test, y_pred))
    print('Test MSE:', metrics.mean_squared_error(y_test, y_pred))
    print('Test RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
    print('test R2:', metrics.r2_score(y_test, y_pred))

```

```python
# Llamada a la función con los conjuntos de datos
regresion(x_train_escaled_concat, x_test_escaled_concat, y_train, y_test)
```

Basándonos en las métricas obtenidas, podemos hacer las siguientes conclusiones:

**Métricas de R-squared (Coeficiente de Determinación):**

La métrica `R-squared` en el modelo sin validación cruzada es bastante alta (aproximadamente 0.95), lo que indica que el modelo explica aproximadamente el 95% de la variabilidad en la variable objetivo a partir de las características proporcionadas.
Sin embargo, al usar validación cruzada, se observa una variabilidad en la métrica R-squared, con valores que van desde 0.459 a 0.934 en diferentes pliegues.

**Métricas en el conjunto de prueba:**

`MAE (Error Absoluto Medio)` y `RMSE (Error Cuadrático Medio)` son métricas que indican la magnitud del error entre las predicciones del modelo y los valores reales en el conjunto de prueba. Valores más bajos indican un mejor rendimiento. En este caso, el MAE y el RMSE son relativamente bajos, lo que sugiere que el modelo tiene un buen desempeño en términos de precisión de predicción en el conjunto de prueba.
`MSE (Error Cuadrático Medio)` también es bajo, lo que indica que el modelo tiene un bajo error cuadrático medio en el conjunto de prueba.

`R2 (Coeficiente de Determinación)` en el conjunto de prueba es aproximadamente 0.8863, lo que sugiere que el modelo explica alrededor del 89% de la variabilidad en la variable objetivo en el conjunto de prueba, como ya lo habiamos estudiado anteriormente.


En resumen, el modelo parece tener un buen rendimiento en la predicción de la variable objetivo en el conjunto de prueba, con un R-squared de alrededor del 89%.
